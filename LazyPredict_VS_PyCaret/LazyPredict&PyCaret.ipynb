# Training Multi Models with Auto-ML LazyPredict and Pycaret

In this study, I will try two python open-source modules which are Pycaret and Lazypredict. Their Classification and Regression task pretty similar. In both you can train your data with a few line codes, and result will occur many models. So, you can get some ideas of your project from results, meaning that which model works better. However, there are some diffiriences between these two modules. As example, LazyPredict doesn't apply any parameter tuning, so you can only train your model with default parameters. This process is faster to get insgiht about to best fitted model. Pycaret seems more robust, because you can train your data with cross-validation at the same time (also basic data preprocessing).Surely that will take more time to train data ..After all, it will be your decision to choose which modules..

- LazyPredict Documentation: https://lazypredict.readthedocs.io.
- Pycaret Documentation: https://pycaret.readthedocs.io/en/latest/index.html

# 1. Lazy Predict
- Note: LazyPredict works better in Python 3.7
- For more information: https://towardsdatascience.com/lazy-predict-fit-and-evaluate-all-the-models-from-scikit-learn-with-a-single-line-of-code-7fe510c7281

```python
# !pip install layzpredict
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris, load_boston #datasets
from sklearn.model_selection import train_test_split
from lazypredict.Supervised import LazyClassifier # For Classification Task
from lazypredict.Supervised import LazyRegressor # For Regression Task
```

## Classification Task

- Firstly, import dataset and train-test split steps..

```python
iris = load_iris()

X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.2,random_state =123)
```

- Only a few line codes you can get the result with metrics ,and get insight which models can fit your data for your future methods.

```python
clf = LazyClassifier(verbose =0, ignore_warnings=True ,predictions=True)
models, predictions = clf.fit(X_train, X_test, y_train, y_test)
models 
```

## Regression Task

```python
boston = load_boston()

X = boston.data
y = boston.target

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.2,random_state =123)
```

```python
reg = LazyRegressor(verbose =0, ignore_warnings=True, predictions=True)
models, predictions = reg.fit(X_train, X_test, y_train, y_test)
models 
```

```python
#before Training your data like that, you have to be sure that correct data types at variables.(numeric or label)
clf = setup(data = train , target = 'species') 
# you can tune attributes list given below in setup part.

# as a default Pycaret apply 10 fold CV (see. description 15-16)
```

## Regression Task

```python
from pycaret.regression import *
```

```python
boston = get_data('boston') # I will choose iris for my Classification tasks

train, test = train_test_split(boston, test_size = 0.2, random_state=123)
```

- Let's set up our models with some data pre-processing methods

```python
reg = setup(data=train, target = 'medv', 
            normalize=True, 
            numeric_imputation = 'mean', 
            remove_outliers=True,
            remove_multicollinearity=True,
            fold_strategy= 'kfold',
            fold =10
           )
#numeric imputation for missing values
#remove_multicollinearity removes less corrolated variable with target variable.
```

```python
best_model = compare_models()
```

```python
GBR = create_model('gbr')
```

```python
GBR_tuned = tune_model(GBR)
```

```python
plot_model(GBR_tuned, plot = 'error')
```

```python
plot_model(GBR_tuned, plot = 'parameter') #model hyperparameters
```

```python
plot_model(GBR_tuned, plot = 'learning')
```

```python
evaluate_model(GBR_tuned)
```

```python
pred_unseen = predict_model(GBR_tuned, data = test) # model testing
```
