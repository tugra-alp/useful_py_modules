# Klib For speed EDA and Data Pre-processing
klib is an easy to use Python library of customized functions for cleaning and analyzing data. You can perform EDA and data preprocessing methods faster and easy way(low-code) with **klib** module.


```python
#!pip install klib
import klib
import sweetviz as sv
import pandas as pd
from pycaret.datasets import get_data # for import datasets
import warnings
warnings.filterwarnings("ignore")
```

```python
data = get_data('titanic')
```

```python
data.info()
```

## 1. Describe functions
### klib.describe functions for visualizing datasets
- klib.cat_plot(df) # returns a visualization of the number and frequency of categorical features
- klib.corr_mat(df) # returns a color-encoded correlation matrix
- klib.corr_plot(df) # returns a color-encoded heatmap, ideal for correlations
- klib.dist_plot(df) # returns a distribution plot for every numeric feature
- klib.missingval_plot(df) # returns a figure containing information about missing values


```python
klib.cat_plot(data=data, figsize=(15,15))
```

```python
klib.corr_mat(data, target = 'Survived') # correlation with response variable
```

```python
# visualized seperatly each negative and positive correlation
klib.corr_plot(data, split='pos', figsize= (12,6))
klib.corr_plot(data, split='neg', figsize= (12,6)) 
```

```python
klib.dist_plot(data) #distribution plots for numeric vars.
```

```python
#Only cabin and age have missing values in dataset
klib.missingval_plot(data, 
                     sort=True, 
                     spine_color = '#ff3333')
# with this function you can see distribution of missing values and statistics values of them(in the right side)
```

## 2. Cleaning Functions

### klib.clean functions for cleaning datasets
- klib.data_cleaning(df) # performs datacleaning (drop duplicates & empty rows/cols, adjust dtypes,...) 
- klib.clean_column_names(df) # cleans and standardizes column names, also called inside data_cleaning()
- klib.convert_datatypes(df) # converts existing to more efficient dtypes, also called inside data_cleaning()
- klib.drop_missing(df) # drops missing values, also called in data_cleaning()
- klib.mv_col_handling(df) # drops features with high ratio of missing vals based on informational content
- klib.pool_duplicate_subsets(df) # pools subset of cols based on duplicates with min. loss of information

```python
data = klib.drop_missing(data,
                  drop_threshold_cols = 0.7
                 ) #as example 'cabin' cols dropped. but 'age' is still there.
```

```python
data.info()
```

```python
#Perform initial data cleaning tasks on a dataset, such as dropping single valued and empty rows, empty columns as well as optimizing the datatypes.
data = klib.data_cleaning(data,
                          drop_duplicates = True,
                          convert_dtypes= True,
                          show = 'all'
                         )
#klib.clean_column_names() and klib.convert_datatypes() are handled in this section.
```

```python
data.info()
```

## 3. Preprocessing Functions
### klib.preprocess functions for data preprocessing (feature selection, scaling, ...)
- klib.train_dev_test_split(df) # splits a dataset and a label into train, optionally dev and test sets
- klib.feature_selection_pipe() # provides common operations for feature selection
- klib.num_pipe() # provides common operations for preprocessing of numerical data
- klib.cat_pipe() # provides common operations for preprocessing of categorical data
- klib.preprocess.ColumnSelector() # selects num or cat columns, ideal for a Feature Union or Pipeline
- klib.preprocess.PipeInfo() # prints out the shape of the data at the specified step of a Pipeline

```python
# Split a dataset and a label column into train, dev and test sets.
X = klib.train_dev_test_split(data,
                              target = 'survived',
                              test_size = 0.2,
                              dev_size= 0,
                              random_state=123
                              ) 
# it returns tuple. containing train-dev-test split of inputs. We have no dev sets.
```
